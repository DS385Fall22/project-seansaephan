# Oh Shit, Git!?!
---------------------

Reference: https://ohshitgit.com/ by Katie Sylor-Miller

"Git is hard: screwing up is easy, and figuring out how to fix your mistakes is fucking impossible."

## Reverting
-------



* Used to get recover stuff you accidentally deleted
* Used to remove some stuff you tried that broke
* Used to recover a bad merge
* Used to go back to a time when things actually worked

## What it looks like in action

Scenario: Made some changes and it broke.



>git reflog



>git reset HEAD@{1}



>git reflog (again)



>git push --force
